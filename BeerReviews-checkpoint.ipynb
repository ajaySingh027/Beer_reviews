# Analyzing the data for Beer reviews :


### Content
Data set is collected from Kaggle. And data contains more than 1.5m reviews for the all the beer worldwide.


### Inspiration

  - Which type (style) of beer is favoured more around the world and which have the top ratings ?
  
  - Which beers have the good quality (review_taste) irrespective of the popularity of beer as number of reviews. ?

  - Does the aroma of the beer is also proportionally related to its higher ratings?

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
```

```python
beer_df = pd.read_csv('beer_reviews.csv')
```

```python
beer_df.head()
```

```python
#shape of beer data
print("Number of reviews: ", beer_df.shape[0])
print("Number of features: ", beer_df.shape[1])
print("Total number of breweries; ", len(beer_df['brewery_name'].value_counts()))
print("Total number of beer style; ", len(beer_df['beer_style'].value_counts()))
print("Total number of beer names; ", len(beer_df['beer_name'].value_counts()))
```

```python
missing_data_columns = beer_df.isnull().mean().sort_values(ascending=False)*100
missing_data_columns.plot(kind='bar')
plt.title('% of misisng data in columns')

missing_data_columns
```

By looking at the above plot, 'beer_abv' columns has the most missing values. And we won't be needin this column values for our analysis. 
All other columns seems to be fine with very minimal or zero missing data.



```python
# drop column beer_abv 
beer_df = beer_df.drop('beer_abv', axis=1)
beer_df.head()
```



### Missing data by rows
Our primary columns are - 'brewery_name', 'beer_name', 'review_taste', 'review_overall', 'review_taste', 'review_aroma' and 'beer_style'.
So removing the rows which have missing data in the listed rows.

```python
# missing data by rows
beer_row_clean = beer_df.dropna(subset=['brewery_name', 'beer_name', 'review_taste', 'review_overall', 'review_taste', 'review_aroma','beer_style'], how='any')
beer_row_clean.head()
```

```python
beer_row_clean.shape
```



### 1. Which type (style) of beer is favoured more around the world and which have the top ratings ?

Now we'll group the beer data as per the beer_style and analyze the ratings (reviews) of the same.

```python
beer_style_count = beer_row_clean.groupby('beer_beerid').agg(dict(beer_style='first', brewery_name ='first', review_overall= 'first', review_aroma='first', review_taste='first', beer_name='first')).reset_index()
beer_style_count
```

```python
beer_style_number = beer_style_count['beer_style'].value_counts()
print("Total number of beer types: ", len(beer_style_number))
beer_style_number
```

Since beer_style list is huge, so we'll consider the top 12 beer types.
Now will see the plot for beers with highest number of reviews having more than 30k reviews overall.
I'll consider more reviews of beer as more consumption around the world.

```python
beer_style_number[:12].plot(kind='bar')
plt.title('Beer with reviews')
plt.xlabel('Beer Types')
plt.ylabel('Number of reviews')

print("Beer with top ratings:----- \n{}".format(beer_style_number[:12]))
```

```python
beer_style_count.head(10)
```



### Which beers have the good quality (review_taste) irrespective of the popularity of beer as number of reviews. ?

In this section we'll analyze the good quality of the beer with more taste reviews given by users worldwide. Here i'm assuming the taste_review as the quality of the beer.

```python
beer_quality = beer_style_count.groupby('beer_style').agg(dict(review_taste='mean')).reset_index()
beer_quality.sort_values('review_taste', ascending=False)
#beer_quality
```

```python
beer_quality = beer_quality[:15]
plt.figure(figsize=(15,8))
sns.barplot(beer_quality['beer_style'], beer_quality['review_taste'], alpha=0.9)
plt.xticks(rotation='vertical')
plt.xlabel("Beer Type", fontsize=12)
plt.ylabel("Taste as Quality", fontsize=12)
plt.title("Beer Types with taste", fontsize=16)
plt.show()
```

#### Analysis:-
If we compare the above two plots as 'Beer with top ratings' vs 'Beer Types with Taste'.
Top 5 brands in first graph: 
(American IPA, American Pale Ale (APA) , American Amber / Red Ale, American Porter, Fruit / Vegetable Beer)

few Top beer brands in Quality wise : --
(American Double / Imperial Stout, American Barleywine, American Black Ale ...)

Both have different listings. Which tells us that even if a famous beer is consumed or prepared by most brewries wordwide, the tastes of other less famous beers can be better in quality or taste.
Quality of beer can be dependent on other features too, such as aroma, appearence, overall review.



### Does the aroma of the beer is also proportionally related to its higher ratings?

In this section, i'll try to figure out if the feature - 'beer aroma' is also a big factor in getting the beer its good reviews.

```python
beer_aroma = beer_style_count.groupby('beer_style').agg(dict(review_aroma='mean')).round(3).reset_index()
beer_aroma.sort_values('review_aroma', ascending=False,)
```

```python
#Create a scatter plot for beer aroma reviews 
beer_aroma = beer_aroma[:30]
#aroma = sns.load_dataset('beer_aroma')
sns.set()
ax = sns.scatterplot(x="beer_style", y=beer_aroma["review_aroma"], hue='review_aroma', size=beer_aroma["review_aroma"], sizes=(20, 200), data=beer_aroma)
ax.set_xticklabels(range(50));
```

```python
# Beer aroma list which is plotted above.
beer_aroma.sort_values('review_aroma', ascending=False,)[:30]
```

For the x-axis of the plot, the names were getting overlapped so have used the numbers to dislpay.
The plot clearly shows the aroma points and only one few of the values are touching the rating 4 whereas the rating goes down with the list.

### Analyzing all major features together to get the best beer
Now lets keep all the major factors together in one datatable to get the best beer to be chosen and analyze how the data deviates from the individual analysis patterns.

```python
beer_style_count['total_reviews'] =0
beer_overall_best = beer_style_count.groupby(['beer_style']).agg(dict(brewery_name='first', total_reviews='count', review_overall='median', review_taste='mean', review_aroma='mean')).reset_index()
beer_overall_best.sort_values(by='total_reviews', ascending=False)
```

### Conclusion
In the final conclusion in the dataset above, we can nicely observe that to taste the best quality beer anywhere around the world then checking highest number of reviews for the beer style or the brewery name, is not the only one point to be considered.
The features which plays the major factor in choosing the best beer are 'Beer Aroma', 'Beer rating' and which directly selects the best tasting beer. :)
